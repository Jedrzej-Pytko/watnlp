```python
frequency_conditional = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
```

```python
genres = ['news', 'religion', 'hobbies', 'science fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
frequency_conditional.tabulate(conditions=genres, samples=modals)
```

```python
def generate_model(cfdist: ConditionalFreqDist, word, num=15):
    for i in range(num):
        print(word, end=' ')
        word = cfdist[word].max()

text = brown.words()
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
```

```python
generate_model(cfd, "")
```
